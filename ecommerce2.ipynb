## Vision Coin v.001 Overview

Well well well... the Vision Coin is doing quite well I must say. Micah would be highly impressed.

 Day 2: Just getting started. Digital value not that well. Marketing was really low.
 
 Day 5: Okay okay... we jumped after marketing to a Digital Value of more than 4 which is great for v.001.

```python
# Let's see the overview of Vision Coin v.001. This will be an overview of the week released. Impressive I must say.

days = [2,5,7]
digital = [1,5,9]



plt.figure(figsize=(15,6))
plt.plot(days, digital, color ='green', linewidth =2, linestyle = '--',label='Vision Coin')
plt.xlabel('Days')
plt.ylabel('Digital Value')
plt.title('Vision Coin v.001')
plt.show()
```

# Alt Coin

Just as we assumed. There is competiton in the market and they have released their charts. Our secetary gave them to me for review. After seeing their chart, we knew we were going to have some exciting news to share. Let's take a look.


```python
# Let's see the overview of Vision Coin v.001.

days = [2,5,7]
digital = [1,8,2]



plt.figure(figsize=(15,6))
plt.plot(days, digital, color ='red', linewidth =8, linestyle = 'solid',label='Alt Coin')
plt.xlabel('Days')
plt.ylabel('Digital Value')
plt.title('Alt Coin v.001')
plt.show()
```

# Vision Coin VS Alt Coin
Let's take a look at both in one viewing. As you can see... the Alt Coin started extremely well. But when they took our A2020 chips, they were exposed and many investors sold their shares of crypto. LOL. 

```python
# Let's see the two on one chart. I think that will help our productivity time. LOL.
# Change width and line colors


days = [2,5,7]
digital = [1,5,9]

daysAlt = [2,5,7]
digitalAlt = [1,8,2]

plt.figure(figsize=(15,6))
plt.plot(days,digital, color ='green', linewidth =8, linestyle = '--',label='Vision Coin')
plt.plot(daysAlt,digitalAlt, color ='red', linewidth =8, linestyle = '-',label='Alt Coin')
plt.xlabel('Days')
plt.ylabel('Digital Value')
plt.title('Vision Coin VS Alt Coin')
plt.legend()
plt.show()
```

# Bar Graph View

- So here's another view of how the week went. The ALt Coin was making major growth witin the beginning. However Vision Coin wasn't focused on the week as a race. They were viewing it as a marathon!

```python

days = [2,5,7]
digital = [1,5,9]

daysAlt = [2,5,7]
digitalAlt = [1,8,2]

plt.figure(figsize=(10,5))
plt.bar(days,digital, color ='green', linewidth =8, linestyle='dashed',label='Vision Coin')
plt.bar(daysAlt,digitalAlt, color ='red', linewidth =8, linestyle = 'dashed',label='Alt Coin')
plt.xlabel('Days')
plt.ylabel('Digital Value')
plt.title('Vision Coin VS Alt Coin')
plt.legend()
plt.show()
```
